```
!wget https://github.com/italojs/facial-landmarks-recognition-/raw/master/shape_predictor_68_face_landmarks.dat
from imutils import face_utils
import dlib
import numpy as np
import cv2
from scipy import io
p = "shape_predictor_68_face_landmarks.dat"
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(p)
in1=input('Enter File Name 1')
gray=cv2.imread(in1)
gray=cv2.cvtColor(gray,cv2.COLOR_RGB2GRAY)
rects=detector(gray,0)
rect=rects[0]
shape=predictor(gray,rect)
shape=face_utils.shape_to_np(shape)
in2=input('Enter File Name 2')
gray1=cv2.imread(in2)
gray1=cv2.cvtColor(gray1,cv2.COLOR_RGB2GRAY)
rects1=detector(gray1,0)
rect1=rects1[0]
shape1=predictor(gray1,rect1)
shape1=face_utils.shape_to_np(shape1)
final=np.concatenate((shape,shape1),axis=1)
io.savemat('tpts.mat',{'tpts':final})
try:
  from google.colab import files
  files.download('./tpts.mat')
except:
  pass
```
